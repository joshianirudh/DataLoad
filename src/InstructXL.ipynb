```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
_20news = fetch_20newsgroups(subset="all")
data = _20news.data
sentences = data
```

```python
_20news.shape
```

```python
from InstructorEmbedding import INSTRUCTOR
model = INSTRUCTOR('hkunlp/instructor-xl')
```

```python
#instruction = "Represent the abstract of the below research paper:"
instruction = "Represent the document from 20NEWS dataset"
```

```python
text = []
for i,sentence in enumerate(sentences):
  text.append([instruction, sentence])
text[0]
```

```python
embeddings = model.encode(text)
```

```python
import numpy as np
np.save("/content/drive/MyDrive/GUILD Embeddings/instruct_xl_20NEWS_embeddings.npy", embeddings)
```
